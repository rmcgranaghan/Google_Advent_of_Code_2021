```python
def calc_winning_score(array,input_num):
    array = array.astype("float")
    array[np.where(array == 999.)] = np.nan
    sum_array = np.nansum(array)
    score = sum_array * input_num
    return score
```

```python
check_boards = np.copy(list_boards_stack)
flag_stop = 0
for num in input_nums:
    print('checking input number = {}'.format(num))
    idx_loop = np.where(list_boards_stack == num)
    check_boards[idx_loop] = 999
    for c in range(len(list_boards_stack)):
        for c2 in range(len(list_boards_stack[c])):
            if (check_boards[c][c2,:] == 999.).all():
                print('Found winning board! = {}'.format(c))
                score = calc_winning_score(check_boards[c],num)
                print('the answer to day 4 part 1 = {}'.format(score))
                flag_stop = 1
                break
            elif (check_boards[c][:,c2] == 999.).all():
                print('Found winning board! = {}'.format(c))
                score = calc_winning_score(check_boards[c],num)
                print('the answer to day 4 part 1 = {}'.format(score))
                flag_stop = 1
                break
            else:
                continue
    if flag_stop == 1:
        break
```

```python
def angle_between(p1, p2):
    ang1 = np.arctan2(*p1[::-1])
    ang2 = np.arctan2(*p2[::-1])
    return np.rad2deg((ang1 - ang2) % (2 * np.pi))
```

```python
def connect(x1,x2,y1,y2):
    ends = np.array(
        [
            [x1,y1],
            [x2,y2]
        ],dtype=int);
#     print(ends)    
    d0, d1 = np.abs(np.diff(ends, axis=0))[0]
    if d0 > d1: 
        return np.c_[np.linspace(ends[0, 0], ends[1, 0], d0+1, dtype=np.int32),
                     np.round(np.linspace(ends[0, 1], ends[1, 1], d0+1))
                     .astype(np.int32)]
    else:
        return np.c_[np.round(np.linspace(ends[0, 0], ends[1, 0], d1+1))
                     .astype(np.int32),
                     np.linspace(ends[0, 1], ends[1, 1], d1+1, dtype=np.int32)]
```
