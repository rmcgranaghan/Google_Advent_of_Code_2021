```python
txt_file = '/Users/ryanmcgranaghan/Documents/Google_Advent_of_Code_2021/day4_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')#.remove('')
f.close()

# int_data = [int(x) for x in txt_data]
# int_data
```

```python
input_nums = txt_data[0]
input_nums = [int(i) for i in input_nums.split(',')]
txt_data = txt_data[1:]
```

```python
txt_data
```

```python
board_number = 0
list_boards = []
for t in range(0,len(txt_data),6):
    print('empty')#txt_data[t])
    array_tmp = np.zeros( (5,5) )
    for r in range(5):
        print('   {}'.format(txt_data[t+r]))
        array_tmp[r,:] = txt_data[t+1+r].split()
    list_boards.append(array_tmp)
```

```python
np.shape(list_boards)
```

```python
list_boards_stack = np.stack(list_boards)
```

```python
def calc_winning_score(array,input_num):
    array = array.astype("float")
    array[np.where(array == 999.)] = np.nan
    sum_array = np.nansum(array)
    score = sum_array * input_num
    return score
```

```python
check_boards = np.copy(list_boards_stack)
flag_stop = 0
for num in input_nums:
    print('checking input number = {}'.format(num))
    idx_loop = np.where(list_boards_stack == num)
    check_boards[idx_loop] = 999
    for c in range(len(list_boards_stack)):
        for c2 in range(len(list_boards_stack[c])):
            if (check_boards[c][c2,:] == 999.).all():
                print('Found winning board! = {}'.format(c))
                score = calc_winning_score(check_boards[c],num)
                print('the answer to day 4 part 1 = {}'.format(score))
                flag_stop = 1
                break
            elif (check_boards[c][:,c2] == 999.).all():
                print('Found winning board! = {}'.format(c))
                score = calc_winning_score(check_boards[c],num)
                print('the answer to day 4 part 1 = {}'.format(score))
                flag_stop = 1
                break
            else:
                continue
    if flag_stop == 1:
        break
```
