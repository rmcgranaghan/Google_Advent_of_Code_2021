# [Day 1](https://adventofcode.com/2020/day/1)

```python
txt_file = '/Users/ryanmcgranaghan/Documents/Google_Advent_of_Code_2021/day1_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')#.remove('')
f.close()

int_data = [int(x) for x in txt_data]
# int_data
```

```python
int_data
```

```python
range(1,len(int_data))
```

```python
ctr = 0
growth_list = ['N/A']
for c in range(1,len(int_data)):
    
#     print(int_data[c])
    if int_data[c]>int_data[c-1]:
        ctr +=1
        growth_list.append('increased')
#         print('    increased')
    elif int_data[c]<int_data[c-1]:
        growth_list.append('decreased')
#         print('    decreased')
    else:
        growth_list.append('no change')
print('the answer to day 1, part 1 is {}'.format(ctr))
```
